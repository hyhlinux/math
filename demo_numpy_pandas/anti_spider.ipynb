```python
import pandas as pd
```

```python
# 取出24h内访问前50的ip.
df = pd.DataFrame(pd.read_csv('auto_test.cvs'))
print("原始数据:\n{}".format(df))
```

```python
# 按访问总数 && 访问域名比 排序
df = df.sort_values(['total', 'download.pureapk.com'], ascending=False)
# 只封100% && 访问次数超过1000次.
df_spider = df[df['download.pureapk.com'] == '100.00%']
df_spider = df_spider[df_spider['total'] > 1000]
print(df_spider)
```
