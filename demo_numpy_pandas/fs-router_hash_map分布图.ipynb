```python
# 查看值域区间
import requests
```

```python
resp = requests.get("http://172.16.1.104:4004/stats?which=app")
data = resp.json()
```

```python
hash_map = data.get("hostGroupMap", {}).get("default", {}).get("HashMap", {})
hash_map
```

```python
# 绘制分布图
sort_x = [int(x) for x in hash_map.keys()]
sort_x = sorted(sort_x)
sort_x
```

```python
hash_map_sort = sorted(hash_map.items(), key=lambda d:int(d[0]))
# y_host_name = [hash_map[str(x)] for x in sort_x]
# print(y_host_name)
# conver_map = {
#     "d-01.winudf.com": 10,
#     "d-02.winudf.com": 20,
#     "d-03.winudf.com": 30,
#     "d-04.winudf.com": 40,
#     "d-05.winudf.com": 50,
# }
# y_host_value = [conver_map[key] for key in y_host_name]
# print(y_host_value)
hash_map_sort
```

```python
# 组装数据
last = 0
def diff(val):
    global last
    ret = val - last
    last = val
    return ret

ret = [(k[1], k[0], diff(int(k[0])))  for k in hash_map_sort]
ret
```

```python
# 计算每个host->diff和
def diff_sum():
    diff_sum_map = {}
    for key in ret:
        diff_sum_map[key[0]] = key[2] + diff_sum_map.get(key[0], 0)
    return diff_sum_map
diff_map = diff_sum()
diff_map
```
